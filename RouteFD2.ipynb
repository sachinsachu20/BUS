```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time

# Initialize WebDriver
driver_k = webdriver.Chrome()

Bus_name_k = []
Bus_type_k = []
Start_time_k = []
End_time_k = []
Rating_k = []
Total_duration_k = []
Prices_k = []
Seats_available_k = []
Route_names_k = []
Route_links_k = []

# Iterate through DataFrame rows
for i, r in df.iterrows():
    link = r["Route_link"]
    routes = r["Route_name"]

    # Loop through each link
    driver_k.get(link)
    time.sleep(2)

    # Click on elements to reveal bus details
    elements = driver_k.find_elements(By.XPATH, "//a[contains(@href,'(link)')]")
    for element in elements:
        element.click()
        time.sleep(2)

        # Click elements to view bus
        try:
            clicks = driver_k.find_element(By.XPATH, '//div[@class="button"]')
            clicks.click()
            time.sleep(2)
        except Exception as e:
            print(f"Exception occurred while clicking button: {e}")
            continue  # Continue to next iteration if an exception occurs

    # Scroll down to ensure all data is loaded
    scrolling = True
    while scrolling:
        old_page_source = driver_k.page_source
        
        # Use ActionChains to perform PAGE_DOWN
        ActionChains(driver_k).send_keys(Keys.PAGE_DOWN).perform()
        time.sleep(5)
        
        new_page_source = driver_k.page_source
        if new_page_source == old_page_source:
            scrolling = False

    # Extract bus details
    try:
        bus_name = driver_k.find_element(By.XPATH, '//div[@class="travels lh-24 f-bold d-color"]').text
    except:
        bus_name = 'N/A'

    try:
        bus_type = driver_k.find_element(By.XPATH, '//div[@class="bus-type f-12 m-top-16 l-color evBus"]').text
    except:
        bus_type = 'N/A'

    try:
        start_time = driver_k.find_element(By.XPATH, '//div[@class="dp-time f-19 d-color f-bold"]').text
    except:
        start_time = 'N/A'

    try:
        end_time = driver_k.find_element(By.XPATH, '//div[@class="bp-time f-19 d-color disp-Inline"]').text
    except:
        end_time = 'N/A'

    try:
        total_duration = driver_k.find_element(By.XPATH, '//div[@class="dur l-color lh-24"]').text
    except:
        total_duration = 'N/A'

    try:
        rating = driver_k.find_element(By.XPATH, '//div[@class="lh-18 rating rat-green "]').text
    except:
        rating = 'No Rating'

    try:
        price = driver_k.find_element(By.XPATH, '//div[@class="column-seven p-right-10 w-15 fl"]').text
    except:
        price = 'N/A'

    try:
        seats = driver_k.find_element(By.XPATH, '//div[@class="column-eight w-15 fl"]').text
    except:
        seats = 'N/A'

    # Append data to respective lists
    Bus_name_k.append(bus_name)
    Route_links_k.append(link)
    Route_names_k.append(routes)
    Bus_type_k.append(bus_type)
    Start_time_k.append(start_time)
    End_time_k.append(end_time)
    Total_duration_k.append(total_duration)
    Rating_k.append(rating)
    Prices_k.append(price)
    Seats_available_k.append(seats)

print("Successfully Completed")

# Close the WebDriver
driver_k.quit()
```

```python
#form list to conver data frame
data = {
    'Bus_name' : Bus_name_k,
    'Bus_type' : Bus_type_k,
    'Start_time' : Start_time_k,
    'End_time' : End_time_k,
    'Total_duration' : Total_duration_k,
    'Price' : Prices_k,
    'Seats_Available' : Seats_available_k,
    'Ratings' : Rating_k,
    'Route_link' : Route_links_k,
    'Route_name' : Route_names_k,
}
df_buses_1 = pd.DataFrame(data)

#convert dataframe to csv
df_buses_1.to_csv(r'E:\FD\kerala_routes.csv', index=False)
df_buses_1
```
