```python
import pandas as pd
import mysql.connector

# Load CSV file into DataFrame
df = pd.read_csv('E:\\final_data\\cleaned_routes.csv')
print("Data loaded successfully.")

# Data cleaning functions
def clean_duration(duration):
    try:
        if pd.isna(duration):
            return '00:00:00'
        hours, minutes = duration.split('h')
        minutes = minutes.replace('m', '')
        return f"{hours.zfill(2)}:{minutes.zfill(2)}:00"
    except:
        return '00:00:00'

def clean_price(price):
    try:
        if pd.isna(price):
            return 0.0
        return float(price.replace('Starts from\nINR ', '').replace(' INR', '').replace(',', ''))
    except:
        return 0.0

def clean_seats(seats):
    try:
        if pd.isna(seats):
            return 0
        return int(seats.split()[0])  # To handle cases like '13 Seats available'
    except:
        return 0

def clean_ratings(rating):
    try:
        if pd.isna(rating) or rating == 'No Rating':
            return 0.0
        return float(rating)
    except:
        return 0.0

# Apply cleaning functions
df['Total_duration'] = df['Total_duration'].apply(clean_duration)
df['Price'] = df['Price'].apply(clean_price)
df['Seats_Available'] = df['Seats_Available'].apply(clean_seats)
df['Ratings'] = df['Ratings'].apply(clean_ratings)

# Fill NaN values with empty strings for text columns
df.fillna({
    'Bus_name': '',
    'Bus_type': '',
    'Start_time': '',
    'End_time': '',
    'Route_link': '',
    'Route_name': ''
}, inplace=True)

print("Data cleaned successfully.")

# Database connection
try:
    conn = mysql.connector.connect(
        host='127.0.0.1',
        user='root',
        password='Krishna3_6_9',
        database='route'
    )
    cursor = conn.cursor()
    print("Database connection successful.")
except mysql.connector.Error as err:
    print(f"Error: {err}")
    exit(1)

# Insert data into MySQL
try:
    for index, row in df.iterrows():
        cursor.execute(
            """
            INSERT INTO bus_routes (
                Bus_name, Bus_type, Start_time, End_time, Total_duration, Price, 
                Seats_Available, Ratings, Route_link, Route_name
            ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
            """,
            (row['Bus_name'], row['Bus_type'], row['Start_time'], row['End_time'], 
             row['Total_duration'], row['Price'], row['Seats_Available'], 
             row['Ratings'], row['Route_link'], row['Route_name'])
        )
    conn.commit()
    print(f"Inserted {len(df)} rows successfully.")
except mysql.connector.Error as err:
    print(f"Error inserting row {index}: {row}")
    print(f"Error: {err}")
finally:
    cursor.close()
    conn.close()
    print("Database connection closed.")
```
