```python
#TRAIAL KEARALA

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
import pandas as pd
import time

# URLs to scrape
state_link = [
    "https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchoketile",
    # Add other URLs as needed
]

# Open the browser
driver = webdriver.Chrome()

def kerala_link_route(url, path):
    driver.get(url)
    time.sleep(3)  # Let the page load
    driver.maximize_window()
    
    # Initialize WebDriverWait
    wait = WebDriverWait(driver, 20)

    LINKS_KERALA = []
    ROUTE_KERALA = []
    page_number = 1

    while True:
        try:
            # Retrieve the route links
            paths = driver.find_elements(By.XPATH, path)
            for link in paths:
                d = link.get_attribute("href")
                if d:
                    LINKS_KERALA.append(d)
                ROUTE_KERALA.append(link.text)

            # Print for debugging
            print(f"Page {page_number}: Found {len(paths)} routes.")
            
            # Wait for the pagination element to be present and click next button
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            next_buttons = pagination.find_elements(By.XPATH, '//div[@class="DC_117_pageTabs"]')
            
            # Find the next button based on page number
            next_button = None
            for btn in next_buttons:
                if btn.text == str(page_number + 1):
                    next_button = btn
                    break
            
            if next_button:
                next_button.click()
                page_number += 1
                time.sleep(3)  # Wait for the page to load
            else:
                print("No more pages to paginate.")
                break

        except (NoSuchElementException, TimeoutException) as e:
            print(f"Error: {str(e)}")
            break

    return LINKS_KERALA, ROUTE_KERALA

# Use the function for the specific URL
url = state_link[0]  # Change as needed
LINKS_KERALA, ROUTE_KERALA = kerala_link_route(url, '//a[@class="route"]')

# Create a DataFrame
df_k = pd.DataFrame({
    "Route_name": ROUTE_KERALA,
    "Route_link": LINKS_KERALA
})

# Display the DataFrame
print(df_k)

# Close the browser
driver.quit()
```
